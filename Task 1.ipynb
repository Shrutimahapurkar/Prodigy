```python
# for ploting bar chart or histogram we have select aspecific years population of specific countries
# so we select year 2020
# filter the data 
df_2020 = df[['Country Name', '2020']].dropna()
```

```python
#select some countries 
countries = ['India', 'China', 'South Africa', 'Brazil', 'Russia']
```

```python
#filtering data for selcted countries
df_countries = df_2020[df_2020['Country Name'].isin(countries)]
```

```python
# ploting bar chart 

plt.figure(figsize=(10,6))
plt.bar(df_countries['Country Name'],df_countries['2020'], color='skyblue')
plt.title('Total population of countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
#x variable rotates with 45 degree
plt.tight_layout()
plt.show()
```

```python
# ploting bar chart  2022

df_2022 = df[['Country Name', '2022']].dropna()
countries = ['India', 'China', 'South Africa', 'Brazil', 'Russia']

df_countries = df_2022[df_2022['Country Name'].isin(countries)]

plt.figure(figsize=(10,6))
plt.bar(df_countries['Country Name'],df_countries['2022'], color='green')
plt.title('Total population of countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
#x variable rotates with 45 degree
plt.tight_layout()
plt.show()
```
